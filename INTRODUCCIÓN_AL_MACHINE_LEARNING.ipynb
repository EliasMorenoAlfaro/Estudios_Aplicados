```python
"""
T1: INTRODUCCIÓN AL MACHINE LEARNING
Conceptos fundamentales y tipos de aprendizaje automático
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification, make_regression, make_blobs
import warnings
warnings.filterwarnings('ignore')

# Configurar visualización
plt.style.use('default')
sns.set_palette("husl")

print("=== INTRODUCCIÓN AL MACHINE LEARNING ===\n")

print("1. ¿QUÉ ES EL MACHINE LEARNING?")
print("-" * 35)

print("""
DEFINICIÓN:
El Machine Learning (Aprendizaje Automático) es una rama de la Inteligencia Artificial
que permite a las computadoras aprender y tomar decisiones a partir de datos,
sin ser programadas explícitamente para cada tarea específica.

ANALOGÍA SIMPLE:
Imagina que quieres enseñar a un niño a reconocer perros:
- Método tradicional: Le das una lista de características (4 patas, cola, ladra, etc.)
- Machine Learning: Le muestras miles de fotos de perros y no-perros,
  y él aprende a reconocerlos por sí mismo.

APLICACIONES COTIDIANAS:
• Recomendaciones de Netflix/Spotify
• Detección de spam en email
• Reconocimiento facial en fotos
• Asistentes virtuales (Siri, Alexa)
• Navegación GPS con tráfico en tiempo real
• Detección de fraude en tarjetas de crédito
""")

print("\n2. TIPOS PRINCIPALES DE MACHINE LEARNING")
print("-" * 40)

print("""
 APRENDIZAJE SUPERVISADO (Supervised Learning):
   • Tenemos datos de entrada (X) y respuestas correctas (y)
   • El algoritmo aprende la relación entre X e y
   • Objetivo: Predecir y para nuevos datos X

   Ejemplos:
   - Predecir precio de casa (X: tamaño, ubicación → y: precio)
   - Clasificar emails (X: contenido → y: spam/no spam)
   - Diagnóstico médico (X: síntomas → y: enfermedad)

 APRENDIZAJE NO SUPERVISADO (Unsupervised Learning):
   • Solo tenemos datos de entrada (X), sin respuestas correctas
   • El algoritmo busca patrones ocultos en los datos
   • Objetivo: Descubrir estructura en los datos

   Ejemplos:
   - Segmentar clientes por comportamiento
   - Detectar anomalías en transacciones
   - Reducir dimensionalidad de datos

 APRENDIZAJE POR REFUERZO (Reinforcement Learning):
   • El algoritmo aprende mediante prueba y error
   • Recibe recompensas o castigos por sus acciones
   • Objetivo: Maximizar recompensas a largo plazo

   Ejemplos:
   - Juegos (AlphaGo, videojuegos)
   - Robots que aprenden a caminar
   - Trading automático
""")

print("\n3. DEMOSTRACIÓN VISUAL DE TIPOS DE ML")
print("-" * 35)

# Crear figura con ejemplos visuales
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Tipos de Machine Learning - Ejemplos Visuales', fontsize=16)

# SUPERVISADO - REGRESIÓN
print("Generando ejemplo de REGRESIÓN (Supervisado)...")
X_reg, y_reg = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)
axes[0, 0].scatter(X_reg, y_reg, alpha=0.6, color='blue')
axes[0, 0].set_title('SUPERVISADO: Regresión\n(Predecir valor continuo)')
axes[0, 0].set_xlabel('Variable X (ej: tamaño casa)')
axes[0, 0].set_ylabel('Variable y (ej: precio)')

# Añadir línea de tendencia
z = np.polyfit(X_reg.flatten(), y_reg, 1)
p = np.poly1d(z)
axes[0, 0].plot(X_reg, p(X_reg), "r--", alpha=0.8, linewidth=2, label='Predicción')
axes[0, 0].legend()

# SUPERVISADO - CLASIFICACIÓN
print("Generando ejemplo de CLASIFICACIÓN (Supervisado)...")
X_class, y_class = make_classification(n_samples=200, n_features=2, n_redundant=0,
                                      n_informative=2, n_clusters_per_class=1, random_state=42)
colors = ['red', 'blue']
for i in range(2):
    mask = y_class == i
    axes[0, 1].scatter(X_class[mask, 0], X_class[mask, 1],
                      c=colors[i], label=f'Clase {i}', alpha=0.6)
axes[0, 1].set_title('SUPERVISADO: Clasificación\n(Predecir categoría)')
axes[0, 1].set_xlabel('Característica 1')
axes[0, 1].set_ylabel('Característica 2')
axes[0, 1].legend()

# NO SUPERVISADO - CLUSTERING
print("Generando ejemplo de CLUSTERING (No Supervisado)...")
X_cluster, y_cluster_true = make_blobs(n_samples=200, centers=3, n_features=2,
                                      random_state=42, cluster_std=1.5)
axes[0, 2].scatter(X_cluster[:, 0], X_cluster[:, 1], alpha=0.6, color='gray')
axes[0, 2].set_title('NO SUPERVISADO: Clustering\n(Encontrar grupos ocultos)')
axes[0, 2].set_xlabel('Característica 1')
axes[0, 2].set_ylabel('Característica 2')

# Mostrar los clusters reales
colors_cluster = ['red', 'blue', 'green']
for i in range(3):
    mask = y_cluster_true == i
    axes[1, 0].scatter(X_cluster[mask, 0], X_cluster[mask, 1],
                      c=colors_cluster[i], label=f'Grupo {i+1}', alpha=0.6)
axes[1, 0].set_title('CLUSTERING: Grupos Descubiertos')
axes[1, 0].set_xlabel('Característica 1')
axes[1, 0].set_ylabel('Característica 2')
axes[1, 0].legend()

# EJEMPLO REAL - DATOS DE CLIENTES
print("Creando ejemplo con datos de clientes...")
np.random.seed(42)
n_customers = 300

# Simular datos de clientes
customer_data = {
    'edad': np.random.normal(40, 15, n_customers),
    'ingresos': np.random.lognormal(10.5, 0.5, n_customers),
    'gastos_mensuales': np.random.normal(2000, 600, n_customers),
    'satisfaccion': np.random.randint(1, 11, n_customers)
}

# Crear variable objetivo (churn) basada en satisfacción
churn_prob = 1 / (1 + np.exp(customer_data['satisfaccion'] - 5))  # Sigmoid
customer_data['churn'] = np.random.binomial(1, churn_prob, n_customers)

df_customers = pd.DataFrame(customer_data)

# Visualizar problema de clasificación real
colors_churn = ['green', 'red']
labels_churn = ['Retiene', 'Abandona']
for i in range(2):
    mask = df_customers['churn'] == i
    axes[1, 1].scatter(df_customers[mask]['satisfaccion'],
                      df_customers[mask]['gastos_mensuales'],
                      c=colors_churn[i], label=labels_churn[i], alpha=0.6)
axes[1, 1].set_title('EJEMPLO REAL: Predicción de Churn\n(¿El cliente se va?)')
axes[1, 1].set_xlabel('Satisfacción (1-10)')
axes[1, 1].set_ylabel('Gastos Mensuales')
axes[1, 1].legend()

# Distribución de variables
axes[1, 2].hist([df_customers[df_customers['churn']==0]['ingresos'],
                df_customers[df_customers['churn']==1]['ingresos']],
               bins=20, alpha=0.7, label=['Retiene', 'Abandona'], color=['green', 'red'])
axes[1, 2].set_title('ANÁLISIS: Ingresos vs Churn')
axes[1, 2].set_xlabel('Ingresos Anuales')
axes[1, 2].set_ylabel('Frecuencia')
axes[1, 2].legend()

plt.tight_layout()
plt.show()

print("\n4. PROCESO TÍPICO DE MACHINE LEARNING")
print("-" * 35)

print("""
 METODOLOGÍA ESTÁNDAR:

1 DEFINIR EL PROBLEMA
   • ¿Qué queremos predecir?
   • ¿Es clasificación o regresión?
   • ¿Qué datos necesitamos?

2 RECOLECTAR Y EXPLORAR DATOS
   • Obtener dataset relevante
   • Análisis exploratorio (EDA)
   • Identificar patrones y problemas

3 PREPARAR LOS DATOS
   • Limpiar datos faltantes
   • Transformar variables
   • Seleccionar características importantes

4 DIVIDIR LOS DATOS
   • Entrenamiento (70-80%): Para enseñar al modelo
   • Prueba (20-30%): Para evaluar el modelo
   • Validación (opcional): Para ajustar parámetros

5 ENTRENAR EL MODELO
   • Seleccionar algoritmo apropiado
   • Ajustar el modelo con datos de entrenamiento
   • Optimizar parámetros

6 EVALUAR EL MODELO
   • Probar con datos no vistos
   • Calcular métricas de desempeño
   • Validar que funciona bien

7 IMPLEMENTAR Y MONITOREAR
   • Poner en producción
   • Monitorear desempeño
   • Reentrenar cuando sea necesario
""")

print("\n5. ALGORITMOS MÁS COMUNES")
print("-" * 25)

algorithms_info = {
    "SUPERVISADO - REGRESIÓN": [
        "• Regresión Lineal: Línea recta que mejor ajusta los datos",
        "• Regresión Polinomial: Curvas para relaciones no lineales",
        "• Random Forest: Combina múltiples árboles de decisión",
        "• Support Vector Regression: Encuentra la mejor línea con margen"
    ],
    "SUPERVISADO - CLASIFICACIÓN": [
        "• K-Nearest Neighbors (KNN): Clasifica según vecinos más cercanos",
        "• Árboles de Decisión: Serie de preguntas sí/no",
        "• Random Forest: Combina múltiples árboles",
        "• Support Vector Machine: Encuentra la mejor frontera entre clases",
        "• Regresión Logística: Probabilidad de pertenecer a una clase"
    ],
    "NO SUPERVISADO": [
        "• K-Means: Agrupa datos en k clusters",
        "• Clustering Jerárquico: Crea árbol de agrupaciones",
        "• DBSCAN: Encuentra clusters de densidad variable",
        "• PCA: Reduce dimensiones manteniendo información importante"
    ]
}

for category, algorithms in algorithms_info.items():
    print(f"\n{category}:")
    for alg in algorithms:
        print(f"  {alg}")

print("\n6. MÉTRICAS DE EVALUACIÓN")
print("-" * 25)

print("""
 PARA REGRESIÓN (predecir números):
• MAE (Error Absoluto Medio): Promedio de errores absolutos
• MSE (Error Cuadrático Medio): Penaliza más los errores grandes
• RMSE (Raíz del MSE): En las mismas unidades que la variable
• R² (Coeficiente de Determinación): % de varianza explicada (0-1)

 PARA CLASIFICACIÓN (predecir categorías):
• Accuracy (Precisión): % de predicciones correctas
• Precision: De los que predije positivos, ¿cuántos eran realmente positivos?
• Recall (Sensibilidad): De los realmente positivos, ¿cuántos detecté?
• F1-Score: Promedio armónico de Precision y Recall
• Matriz de Confusión: Tabla detallada de aciertos y errores
""")

print("\n7. EJEMPLO PRÁCTICO SIMPLE")
print("-" * 25)

# Ejemplo muy simple de clasificación
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

print("PROBLEMA: Predecir si un cliente comprará un producto")
print("DATOS: Edad y salario del cliente")

# Crear datos de ejemplo
np.random.seed(42)
n_samples = 200

# Generar datos sintéticos
ages = np.random.randint(18, 65, n_samples)
salaries = np.random.randint(20000, 100000, n_samples)

# Crear regla simple: compra si edad > 30 Y salario > 50000 (con algo de ruido)
purchases = []
for age, salary in zip(ages, salaries):
    if age > 30 and salary > 50000:
        # 80% probabilidad de comprar
        purchase = np.random.choice([0, 1], p=[0.2, 0.8])
    else:
        # 20% probabilidad de comprar
        purchase = np.random.choice([0, 1], p=[0.8, 0.2])
    purchases.append(purchase)

# Crear DataFrame
simple_df = pd.DataFrame({
    'edad': ages,
    'salario': salaries,
    'compra': purchases
})

print(f"\nDataset creado: {len(simple_df)} clientes")
print("Primeras filas:")
print(simple_df.head())

print(f"\nDistribución de compras:")
print(simple_df['compra'].value_counts())
print(f"Porcentaje que compra: {simple_df['compra'].mean()*100:.1f}%")

# Visualizar los datos
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
colors = ['red', 'green']
labels = ['No Compra', 'Compra']
for i in range(2):
    mask = simple_df['compra'] == i
    plt.scatter(simple_df[mask]['edad'], simple_df[mask]['salario'],
               c=colors[i], label=labels[i], alpha=0.6)
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.title('Datos de Clientes: ¿Quién Compra?')
plt.legend()

# Entrenar modelo simple
X = simple_df[['edad', 'salario']]
y = simple_df['compra']

# Dividir datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar modelo KNN
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)

# Hacer predicciones
y_pred = model.predict(X_test)

# Evaluar
accuracy = accuracy_score(y_test, y_pred)

print(f"\n RESULTADOS DEL MODELO:")
print(f"Datos de entrenamiento: {len(X_train)} clientes")
print(f"Datos de prueba: {len(X_test)} clientes")
print(f"Precisión del modelo: {accuracy:.2%}")

# Visualizar predicciones
plt.subplot(1, 2, 2)
colors_pred = ['red', 'green']
labels_pred = ['Predicción: No Compra', 'Predicción: Compra']
for i in range(2):
    mask = y_pred == i
    plt.scatter(X_test[mask]['edad'], X_test[mask]['salario'],
               c=colors_pred[i], label=labels_pred[i], alpha=0.6, marker='s')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.title(f'Predicciones del Modelo\n(Precisión: {accuracy:.1%})')
plt.legend()

plt.tight_layout()
plt.show()

print("\n8. CONCEPTOS CLAVE PARA RECORDAR")
print("-" * 30)

key_concepts = [
    " ML permite a las máquinas aprender de datos sin programación explícita",
    " Supervisado: tenemos respuestas correctas para entrenar",
    " No supervisado: buscamos patrones ocultos sin respuestas",
    " Clasificación: predecir categorías (spam/no spam, compra/no compra)",
    " Regresión: predecir números (precio, temperatura, ventas)",
    " Siempre dividir datos: entrenamiento para aprender, prueba para evaluar",
    " Métricas nos dicen qué tan bien funciona nuestro modelo",
    " ML es un proceso iterativo: probar, evaluar, mejorar"
]

for concept in key_concepts:
    print(f"  {concept}")

print("\n" + "="*50)
print("¡INTRODUCCIÓN AL MACHINE LEARNING COMPLETADA!")
print("="*50)
print("\nAhora estás listo para profundizar en algoritmos específicos.")
print("En los siguientes scripts aprenderemos:")
print("• Regresión lineal paso a paso")
print("• Clasificación con KNN y árboles de decisión")
print("• Evaluación detallada de modelos")
print("• Clustering con K-means")
print("• Proyectos integradores completos")

# Guardar datos para siguientes scripts
simple_df.to_csv('datos_ejemplo_ml.csv', index=False)
print(f"\n Datos guardados en 'datos_ejemplo_ml.csv' para próximos ejercicios")
```
